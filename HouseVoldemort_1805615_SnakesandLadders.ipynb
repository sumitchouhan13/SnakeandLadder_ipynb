```python
ladder = {3 : 38, 24 : 33, 42 : 93, 72 : 84} #positions of ladders in the game
def checkladders(ladder , key):
    if key in ladder.keys():
        print("*****Great!! You jumped a ladder*****")
        return ladder[key]
    else:
        return key
```

```python
snakes = {17 : 7, 54 : 34, 62 : 19, 98 : 79} #positions of snakes in the game
def checksnakes(snakes , key):
    if key in snakes.keys():
        print("*****Oh!! You encountered a snake*****")
        return snakes[key]
    else:
        return key
```

```python
def valueschecker(): #This function is to check values for auto mode
    roll_1 = input()
    if roll_1 == 'roll':
        val1 = random.randint(1 , 6)
        return val1
    elif roll_1 == 'quit':
        return 0
    else:
        print("Wrong input , try again")
        return valueschecker()
```

```python
def valuesmanualchecker(): #This function is to check values for manual mode
    #print("Enter a number from 1 to 20 : ")
    roll_1 = int(input("Enter a number from 1 to 20 : "))
    if roll_1 >= 1 and roll_1 <= 20:
        val1 = roll_1
        return val1
    elif roll_1 == 0:
        return 0
    else:
        print("Wrong input , Please enter a number between 1 to 20")
        return valuesmanualchecker()
```
